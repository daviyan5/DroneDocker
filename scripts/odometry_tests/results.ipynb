```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pickle

# Load the pickle file
with open('./results/finalized_20231008-043644.pkl', 'rb') as file:
    iterations_list, last_iteration = pickle.load(file)

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(iterations_list)
df_copy = df.copy()
# In df_copy, create the rmse columns and remove the timestamps, cpu_usage, memory_usage and poses columns
df_copy['cpu_rmse'] = df_copy['cpu_usage'].apply(lambda x: np.sqrt(np.mean(np.square(x))))

df_copy['memory_rmse'] = df_copy['memory_usage'].apply(lambda x: np.sqrt(np.mean(np.square(x))))
# remove timestamps, cpu_usage, memory_usage, and poses columns
df_copy = df_copy.iloc[:, [0, 1, 2, 3, 4, 5, -2, -1]]
print(last_iteration)
# Save the DataFrame to a CSV file
df_copy.to_csv('output.csv', index=False)
```

```python
df_failed = df[df['fail_number'] > 0 | (df['timestamps'].apply(lambda x: len(x) == 0)) | (df['cpu_usage'].apply(lambda x: len(x) == 0)) | (df['memory_usage'].apply(lambda x: len(x) == 0))]
df = df[(df['fail_number'] == 0) & (df['timestamps'].apply(lambda x: len(x) > 0)) & (df['cpu_usage'].apply(lambda x: len(x) > 0)) & (df['memory_usage'].apply(lambda x: len(x) > 0))]
df
```

```python
# Assuming df contains the necessary columns like 'timestamps', 'cpu_usage', 'memory_usage', and 'poses'
for index, row in df.iterrows():
    iteration_number = row['iteration_index']
    timestamps = row['timestamps']
    cpu_usage = row['cpu_usage']
    memory_usage = row['memory_usage']
    poses = row['poses']

    cpu_rmse = np.sqrt(np.mean(np.square(cpu_usage)))
    memory_rmse = np.sqrt(np.mean(np.square(memory_usage)))

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot Trajectory in 3D
    ax.plot(poses[:, 0], poses[:, 1], poses[:, 2], label='Trajectory')

    # Set labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Set title
    ax.set_title(f'Iteration {iteration_number} - CPU RMSE: {cpu_rmse:.3f} - Memory RMSE: {memory_rmse:.3f}')

    # Save the figure
    plt.tight_layout()
    plt.savefig(f'results/images/iteration_{iteration_number}.png')
    plt.close()
```

```python
import os
from IPython.display import display, Image, clear_output

# List to store accepted iteration numbers
accepted_iterations = []

# Directory containing the images
image_dir = 'results/images/'

# Loop through the images in the directory
for filename in os.listdir(image_dir):
    if filename.endswith(".png"):
        iteration_number = int(filename.split('_')[-1].split('.')[0])
        image_path = os.path.join(image_dir, filename)

        # Display the image in the notebook
        display(Image(filename=image_path))

        # Prompt user for input
        user_input = input(f"Accept iteration {iteration_number}? (Y/N): ")
        
        # Clear the output after input
        clear_output(wait=True)

        # Process user input
        if user_input.lower() == 'y':
            accepted_iterations.append(iteration_number)
        elif user_input.lower() == 'n':
            os.remove(image_path)
        else:
            print("Invalid input. Skipping...")

# Print accepted iteration numbers
print(f"Accepted Iteration Numbers: {accepted_iterations}")
```
